## Import libraries


```python
import xarray as xr
import pandas as pd
import numpy as np
import os, glob, sys

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

## Read netcdf file and check it

```python
ds = xr.open_dataset('SST_2010_data.nc')
ds
```

## get one timeslot and quick plot

```python
data = ds.PMSL[1]
data.plot()
```

## plot countour with matplotlib

```python
# Create a grid of latitude and longitude points
lon_grid, lat_grid = data.lon2d, data.lat2d
lon_grid = np.where(lon_grid < 0, lon_grid + 360, lon_grid)
pressure_data = ds.PMSL[1] / 100
```

```python
plt.contourf(lon_grid, lat_grid, pressure_data)
```

## plot countour with cartopy

```python
# Set up the Lambert Conformal Conic projection centered on Japan
fig = plt.figure(figsize=(10, 8))
projection = ccrs.LambertConformal(central_longitude=143, central_latitude=35, standard_parallels=(30, 60))
ax = plt.axes(projection=projection)
ax.contourf(lon_grid, lat_grid, pressure_data, cmap="coolwarm", transform=ccrs.PlateCarree())
```


    ### Step 1: Importing Necessary Libraries / 必要なライブラリのインポート

    In this section, we import essential libraries for data manipulation and plotting.
    - `matplotlib.pyplot` is a standard library for plotting in Python.
    - `cartopy` allows us to work with geographic projections and add map features.
    - `cartopy.feature` provides additional map elements, such as coastlines and borders, for enhanced context.

    These libraries set up the environment for visualizing geographic data.

    ---

    このセクションでは、データ操作とプロットに必要なライブラリをインポートします。
    - `matplotlib.pyplot` はPythonでの標準的なプロットライブラリです。
    - `cartopy` は地理的な投影法の操作と地図の特徴を追加する機能を提供します。
    - `cartopy.feature` は海岸線や国境などの追加の地図要素を提供し、地図のコンテキストを強化します。

    これらのライブラリは、地理データの可視化のための環境を整えます。
    

```python

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Import libraries for plotting and geographic projections
# プロットと地理的な投影法のためのライブラリをインポートします
    
```


    ### Step 2: Setting Up the Lambert Conformal Conic Projection / ランバート正角円錐図法の設定

    In this section, we set up a figure and Lambert Conformal Conic projection centered on Japan.
    - We create a figure with a defined size.
    - The projection is specified with `central_longitude` and `central_latitude` to center it on Japan.
    - `standard_parallels` are defined to minimize distortion in the mid-latitude region, making it ideal for Japan.

    ---

    このセクションでは、日本を中心としたランバート正角円錐図法で図を設定します。
    - 定義されたサイズで図を作成します。
    - 投影法には、日本を中心とするように `central_longitude` と `central_latitude` が指定されています。
    - `standard_parallels` は中緯度地域での歪みを最小限に抑えるために定義されており、日本に適しています。
    

```python

# Set up the Lambert Conformal Conic projection centered on Japan
# 日本を中心としたランバート正角円錐図法を設定します
fig = plt.figure(figsize=(10, 8))

# Lambert Conformal Conic projection with specified central longitude, latitude, and standard parallels
# 中心経度、緯度、標準緯線を指定したランバート正角円錐図法
projection = ccrs.LambertConformal(central_longitude=140, central_latitude=35, standard_parallels=(30, 60))

# Create an axis with the specified projection
# 指定された投影法で軸を作成します
ax = plt.axes(projection=projection)
    
```


    ### Step 3: Adding Map Features for Geographic Context / 地理的コンテキストを提供するための地図の特徴の追加

    We add map features such as coastlines, borders, and land shading.
    - `ax.coastlines()` adds coastlines at a medium resolution.
    - `ax.add_feature(cfeature.BORDERS, linestyle=':')` adds country borders with a dotted style.
    - `ax.add_feature(cfeature.LAND, facecolor='lightgray', edgecolor='black')` shades land areas in light gray for visual context.

    These elements enhance the readability of the map.

    ---

    海岸線、国境、陸地の色付けなどの地図の特徴を追加します。
    - `ax.coastlines()` は中解像度で海岸線を追加します。
    - `ax.add_feature(cfeature.BORDERS, linestyle=':')` は点線スタイルで国境を追加します。
    - `ax.add_feature(cfeature.LAND, facecolor='lightgray', edgecolor='black')` は視覚的なコンテキストのために陸地を薄い灰色で色付けします。

    これらの要素は地図の読みやすさを向上させます。
    

```python

# Add map features for better context
# 地理的なコンテキストを明確にするために地図の特徴を追加します
ax.coastlines(resolution='50m')  # Adds coastlines
ax.add_feature(cfeature.BORDERS, linestyle=':')  # Adds country borders with dotted line
ax.add_feature(cfeature.LAND, facecolor='lightgray', edgecolor='black')  # Shades land areas
    
```
