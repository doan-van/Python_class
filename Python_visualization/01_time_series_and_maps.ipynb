### Extracting Data and Creating a Time Series Plot

This code extracts the year (x-axis) and values (y-axis) from the DataFrame and creates a time series plot. The data is plotted in light gray, with thin lines and some transparency.

- **`x = df.index`**: Extracts the DataFrame’s index (typically the years) to use as x-axis values.

### 年間平均気温上昇の計算とプロット

### Calculating and Plotting Yearly Average Temperature Increase

This code calculates the yearly average temperature increase across all locations and plots it with a highlighted style.

- **`yearly_avg_temp_increase = df.mean(axis=1)`**: Calculates the yearly average temperature increase by taking the mean across all columns in the DataFrame.

### Setting Map Extent and Title for a Map of Japan

This code uses the PlateCarree projection to display a map focused on Japan, setting the map extent and adding a title.

- **`ax.coastlines(resolution='10m')`**: Adds high-resolution coastlines to emphasize Japan’s geographic features.
- **`ax.set_extent(map_extent, crs=ccrs.PlateCarree())`**: Sets the map extent to `[122, 153, 24, 46]`, covering Japan’s area.
- **`plt.title("Map of Japan")`**: Adds the title "Map of Japan" to clarify the map’s content.



```python
plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines(resolution='10m')
#===========
# Set extent to cover Japan [longitude_min, longitude_max, latitude_min, latitude_max]
map_extent = [122, 153, 24, 46]
ax.set_extent(map_extent, crs=ccrs.PlateCarree())

# Set the title of the map to "Map of Japan"
plt.title("Map of Japan")
```

### 日本地図に地理要素を追加 

PlateCarree投影を使用して日本地図を表示し、国境、陸地、海、湖、川などの地理要素を追加します。

- **`ax.set_extent(map_extent, crs=ccrs.PlateCarree())`**: 日本をカバーする表示範囲を設定します。
- **`ax.add_feature(cfeature.BORDERS, linestyle=':')`**: 点線スタイルで国境を追加し、地理的な位置関係を示します。
- **`ax.add_feature(cfeature.LAND, facecolor='lightgray')`**: 陸地を薄い灰色で追加し、地図に地上のエリアを強調します。
- **`ax.add_feature(cfeature.OCEAN, facecolor='lightblue')`**: 海を薄い青色で表示し、水域を明示します。
- **`ax.add_feature(cfeature.LAKES, facecolor='blue', alpha=0.5)`**: 青色と少しの透明度で湖を追加します。
- **`ax.add_feature(cfeature.RIVERS)`**: 川をデフォルトスタイルで追加し、水の流れを表示します。
- **`ax.gridlines()`**: 読みやすさを向上させるためにグリッド線を追加し、必要に応じてラベルを非表示にします。

### Adding Geographic Features to the Map of Japan

This code displays a map of Japan using the PlateCarree projection, adding geographic features such as borders, land, oceans, lakes, and rivers.

- **`ax.set_extent(map_extent, crs=ccrs.PlateCarree())`**: Sets the map extent to cover Japan.
- **`ax.add_feature(cfeature.BORDERS, linestyle=':')`**: Adds country borders with a dotted line style to show geopolitical boundaries.
- **`ax.add_feature(cfeature.LAND, facecolor='lightgray')`**: Adds land in light gray to emphasize land areas.
- **`ax.add_feature(cfeature.OCEAN, facecolor='lightblue')`**: Displays oceans in light blue, clearly indicating water regions.
- **`ax.add_feature(cfeature.LAKES, facecolor='blue', alpha=0.5)`**: Adds lakes with a blue color and some transparency.
- **`ax.add_feature(cfeature.RIVERS)`**: Adds rivers in default style to show water flow.
- **`ax.gridlines()`**: Adds optional gridlines to improve readability, with labels hidden on the top and right edges if not needed.

```python

# Initialize a figure with PlateCarree projection directly
plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent(map_extent, crs=ccrs.PlateCarree())
ax.coastlines(resolution='10m')
plt.title("Map of Japan")

#=======
# Add country borders with a dotted line style
ax.add_feature(cfeature.BORDERS, linestyle=':')
# Add land with a light gray color
ax.add_feature(cfeature.LAND, facecolor='lightgray')
# Add oceans with a light blue color
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
# Add lakes with a blue color and some transparency
ax.add_feature(cfeature.LAKES, facecolor='blue', alpha=0.5)
# Add rivers in default style
ax.add_feature(cfeature.RIVERS)
# Optional gridlines for better readability
gl = ax.gridlines(draw_labels=True, linestyle='--', color='gray')
gl.top_labels = False  # Hide labels on the top edge
gl.right_labels = False  # Hide labels on the right edge
```

### Marking Tokyo on the Map

This code displays a map of Japan using the PlateCarree projection and marks Tokyo with a red dot and label.

- **`ax.plot()`**: Plots a red dot to mark Tokyo’s location.
- **`ax.text()`**: Adds the label "Tokyo" next to the red dot to indicate the city’s name.



```python
plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent(map_extent, crs=ccrs.PlateCarree())

# Add coastlines and other features
ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.add_feature(cfeature.LAKES, facecolor='blue', alpha=0.5)
ax.add_feature(cfeature.RIVERS)

# Optional gridlines
gl = ax.gridlines(draw_labels=True, linestyle='--', color='gray')
gl.top_labels = False
gl.right_labels = False

#======================
# Coordinates for Tokyo
tokyo_lon, tokyo_lat = 139.6917, 35.6895
# Mark Tokyo with a red dot and label
ax.plot(tokyo_lon, tokyo_lat, marker='o', color='red', markersize=5, transform=ccrs.PlateCarree())
# Add a label "Tokyo" next to the red dot
ax.text(tokyo_lon + 0.5, tokyo_lat, 'Tokyo', transform=ccrs.PlateCarree(), fontsize=20, color='red')
plt.title("Map of Japan")
```

```python

plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent(map_extent, crs=ccrs.PlateCarree())
# Add coastlines and other features
ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.add_feature(cfeature.LAKES, facecolor='blue', alpha=0.5)
ax.add_feature(cfeature.RIVERS)
# Optional gridlines
gl = ax.gridlines(draw_labels=True, linestyle='--', color='gray')
gl.top_labels = False
gl.right_labels = False
plt.title("Map of Japan with Colored Weather Stations")

#==========
# Normalize the color scale based on the range of point values
norm = mcolors.Normalize(vmin=min(point_values), vmax=max(point_values))
# Set the color map to 'hot' for a gradient from black to red to yellow
color_map = cm.hot
# Plot points for each station with color based on station_id in a single scatter plot
sc = ax.scatter(longitude, latitude, c=point_values, cmap=color_map, s=20, norm=norm, transform=ccrs.PlateCarree())

```

### Creating and Customizing the Colorbar Placement

This code creates a colorbar based on the colormap and customizes its placement on the map.

- **`plt.cm.ScalarMappable(cmap=color_map, norm=norm)`**: Creates a `ScalarMappable` based on the colormap and normalized scale, which will be used for the colorbar.
- **`sm.set_array([])`**: Sets an array for the `ScalarMappable`, though an empty array is used here for visual purposes.
- **`fig.colorbar()`**: Adds the colorbar to the figure with specific placement properties, including `orientation='horizontal'` for horizontal placement, `pad=0.05` to adjust its position, `shrink=0.5` to reduce size, and `anchor=(0.8, 0.5)` for custom positioning.
- **`cbar.set_label()`**: Sets the label on the colorbar to "Temperature ($^oC$)" to indicate the temperature data.



```python

fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent(map_extent, crs=ccrs.PlateCarree())
# Add coastlines and other features
ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.add_feature(cfeature.LAKES, facecolor='blue', alpha=0.5)
ax.add_feature(cfeature.RIVERS)
# Optional gridlines
gl = ax.gridlines(draw_labels=True, linestyle='--', color='gray')
gl.top_labels = False
gl.right_labels = False
plt.title("Map of Japan with Colored Weather Stations")

norm = mcolors.Normalize(vmin=min(point_values), vmax=max(point_values))
color_map = cm.hot
sc = ax.scatter(longitude, latitude, c=point_values, cmap=color_map, s=20, norm=norm, transform=ccrs.PlateCarree())

#==========
# Create the colorbar and place it inside the map
sm = plt.cm.ScalarMappable(cmap=color_map, norm=norm)
sm.set_array([])  
# Add the colorbar to the figure with custom placement and appearance
cbar = fig.colorbar(sm, ax=ax, orientation='horizontal',
                    pad=0.05, fraction=0.05, shrink=0.5, anchor=(0.8, 0.5))

# Set the label for the colorbar
cbar.set_label("Temperature ($^oC$)")

```
