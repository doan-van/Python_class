## Pythonの基本的な演算

```python
# 2つの数を足す
a = 5
b = 3
result = a + b
print("足し算:", result)  # 出力: 足し算: 8

# 2つの数を引く
a = 10
b = 4
result = a - b
print("引き算:", result)  # 出力: 引き算: 6
```

```python

# 2つの数を掛ける
a = 7
b = 6
result = a * b
print("掛け算:", result)  # 出力: 掛け算: 42

# 1つの数を別の数で割る
a = 20
b = 5
result = a / b
print("割り算:", result)  # 出力: 割り算: 4.0

# 割り算の余りを求める
a = 17
b = 3
result = a % b
print("剰余:", result)  # 出力: 剰余: 2
```

### Pythonの配列（リスト）
Pythonでは、配列の代わりに「リスト」というデータ構造を使用します。リストは、複数の要素を順番に格納できる可変長のデータ構造です。

1. リストの作成
リストは、角括弧 [] を使って作成します。

説明: この例では、3つの果物の名前を含むリストを作成しています。


```python
# リストの作成
fruits = ["りんご", "バナナ", "オレンジ"]
print("果物リスト:", fruits)  # 出力: 果物リスト: ['りんご', 'バナナ', 'オレンジ']
```

2. リストの要素へのアクセス
リストの要素には、インデックスを使用してアクセスします。インデックスは0から始まります。

説明: インデックス 0 を使ってリストの最初の要素（りんご）にアクセスしています。

```python
# リストの要素へのアクセス
first_fruit = fruits[0]
print("最初の果物:", first_fruit)  # 出力: 最初の果物: りんご
```

# 数値の配列
NumPy（Numerical Python）とは
NumPyはPythonの数値計算ライブラリで、高性能な数値計算を効率的に行うための機能を提供します。
特に、大規模な配列や行列の操作が得意で、数学的な計算を高速に行うことができます。

NumPyのインポート
NumPyを使用する際は、通常 import 文を使ってインポートします。


```python
import numpy as np
```

```python
array_1d = np.array([1, 2, 3, 4])
array_2d = np.array([[1, 2], [3, 4]])
print("1次元配列:", array_1d)  # 出力: 1次元配列: [1 2 3 4]
print("2次元配列:\n", array_2d)  # 出力: 2次元配列: [[1 2]
```

配列の属性
NumPy配列は、サイズや形状を持っています。

```python
print("1次元配列の形状:", array_1d.shape)  # 出力: 1次元配列の形状: (4,)
print("2次元配列の形状:", array_2d.shape)  # 出力: 2次元配列の形状: (2, 2)
```

### 配列の演算

```python
array_1d = np.array([1, 2, 3])
result_add = array_1d + 5
result_mult = array_1d * 2
print("配列に5を加算:", result_add)  # 出力: 配列に5を加算: [6 7 8]
print("配列を2倍:", result_mult)      # 出力: 配列を2倍: [2 4 6]
```

### 配列のスライシング

```python

sliced_array = array_1d[1:3]
print("スライスした配列:", sliced_array)  # 出力: スライスした配列: [2 3]
```

Pythonのforループとif文
1. forループ
forループは、リストやタプル、文字列などの要素を1つずつ取り出して繰り返し処理を行うための構文です。

1.1. 基本的な構文

```python
fruits = ["りんご", "バナナ", "オレンジ"]
for fruit in fruits:
    print(fruit)
```

1.2. インデックスを使用する
range()関数を使うと、インデックスを使ってループを回すこともできます。

```python
# インデックスを使ったループ
for i in range(5):
    print("番号:", i)
```

2. if文
if文は、条件に基づいて異なる処理を実行するための構文です。

2.1. 基本的な構文

```python
# 条件に基づいて処理を分岐
number = 10
if number > 5:
    print("5より大きい")
```

2.2. elseとelif
elseやelifを使うことで、複数の条件をチェックできます。


```python
# 複数の条件をチェック
number = 5
if number > 5:
    print("5より大きい")
elif number == 5:
    print("5と等しい")
else:
    print("5より小さい")
```

3. forループとif文の組み合わせ
forループとif文を組み合わせて、特定の条件を満たす要素に対して処理を行うことができます。


```python
# リストの中の特定の条件に合う要素をチェック
numbers = [1, 2, 3, 4, 5]
for number in numbers:
    if number % 2 == 0:
        print(number, "は偶数です")  # 偶数の場合
    else:
        print(number, "は奇数です")  # 奇数の場合
```
